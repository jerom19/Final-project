KMeans Clustering

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaled = StandardScaler().fit(data)       
features = scaled.transform(data)
```

```python
features_scaled = pd.DataFrame( features, columns= data.columns[0:38])
features_scaled.head()
```

Creating clusters from the range of 3 to 15 and find the correct cluster using silhouette_scor

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
for i in range(3,15):
    kmeans = KMeans(n_clusters=i, max_iter=50)
    kmeans.fit(features_scaled)
    cluster_labels = kmeans.labels_
    silhouette_avg = silhouette_score(features_scaled, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(i, silhouette_avg))
```

Build the model with K=4

```python
from sklearn.cluster import KMeans
Km_cluster=KMeans(n_clusters=3,random_state=10)
Km_cluster.fit(features_scaled)

```

```python
pred=Km_cluster.fit_predict(features_scaled)
```

```python
cluster=data.copy()
```

```python
cluster['Cluster']=Km_cluster.predict(features_scaled)
#Loading the predicted values in new column
```

```python
cluster
```

```python
np.unique(Km_cluster.labels_, return_counts=True)                
```

Cluster Visualization

```python
sns.countplot(data= cluster, x = 'Cluster')
plt.title('Cluster Sizes', fontsize = 20)
plt.xlabel('Clusters', fontsize = 15)
plt.ylabel('No. of Customers', fontsize = 15)
plt.text(x = -0.18, y =2000, s = np.unique(Km_cluster.labels_, return_counts=True)[1][0])
plt.text(x = 0.9, y =2000, s = np.unique(Km_cluster.labels_, return_counts=True)[1][1])
plt.text(x = 1.85, y =2000, s = np.unique(Km_cluster.labels_, return_counts=True)[1][2])

plt.show()
```

```python
plt.scatter(cluster.iloc[pred==0,0],cluster.iloc[pred==0,1],s=50,c='black',label='cluster 1')
plt.scatter(cluster.iloc[pred==1,0],cluster.iloc[pred==1,1],s=50,c='red',label='cluster 2')
plt.scatter(cluster.iloc[pred==2,0],cluster.iloc[pred==2,1],s=50,c='pink',label='cluster 3')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=50,c='blue')
plt.legend()
```
